#### Convert date column to datetime

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

#### Days with most strikes

```python
df.groupby(['date']).sum().drop(columns='center_point_geom',errors='ignore').sort_values('number_of_strikes',ascending=False).head(10)
```

#### Create new month column

```python
df['month'] = df['date'].dt.month
df.head()
```

```python
df.groupby(['month']).sum(numeric_only=True).sort_values('number_of_strikes',ascending=False).head(12)
```

#### Convert month to txt

```python
df['month_txt'] = df['date'].dt.month_name().str.slice(stop=3)
df.head()
```

#### Goal is to print total number of strikes per month

```python
df_by_month = df.groupby(['month','month_txt']).sum(numeric_only=True).sort_values('month', ascending=True).head(12).reset_index()
df_by_month
```
